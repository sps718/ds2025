##### Example: Simple Recursive Function

A recursive function is a function that calls itself. Here’s a very basic example that counts down from a number to 1:

```python
def countdown(n):
    if n == 0:
        print("Done!")
    else:
        print(n)
        countdown(n - 1)

# Example usage:
countdown(3)
# Output:
# 3
# 2
# 1
# Done!
```

Notice how `countdown` calls itself with `n - 1` until it reaches the base case (`n == 0`).  
This is the basic idea behind recursion:  
- We create an `if/elif/else` code block to handle different cases.
- We **establish a base case**—an "exit" condition that stops the recursion (here, when `n == 0`).
- For all other cases, we perform some action and then **call the function itself** with a new argument that brings us closer to the base case.

By always moving closer to the base case, we ensure that the recursion will eventually stop. This pattern is used in many classic problems, such as calculating Fibonacci numbers or factorials.
