```python
import matplotlib.pyplot as plt
```

```python

# explicit demo

fig, ax = plt.subplots()

ax.plot([1, 2, 3, 4, 3, 5, 2, 4])
ax.set_title("Test Plot")
ax.set_xlabel("x-axis")
ax.set_ylabel("y-axis")

fig.savefig("test.png")
```

```python
# implicit demo

plt.plot([1,2,3,4,3,5,2,4])
plt.title("Test Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")

plt.savefig("test2.png")
plt.show()
```
